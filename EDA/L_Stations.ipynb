```python
df = pd.read_pickle(os.path.join(SERIALIZED, "CTA_-_System_Information_-_List_of__L__Stops.pkl"))
```

```python
# format geolocation
for idx, row in df.iterrows():
    loc_split = row.Location.split(", ")
    lon = loc_split[1].strip("()")
    lat = loc_split[0].strip("()")
    df.loc[idx, "LATITUDE"] = lat
    df.loc[idx, "LONGITUDE"] = lon

# fix data types
df["LATITUDE"] = df["LATITUDE"].astype(float)
df["LONGITUDE"] = df["LONGITUDE"].astype(float)
```

```python
# drop unnecessary columns
df.drop([
    "STOP_ID",
    "DIRECTION_ID",
    "STOP_NAME",
    "STATION_DESCRIPTIVE_NAME",
    "MAP_ID"
], axis=1, inplace=True)
```
