```python
df_grocery = pd.read_pickle(os.path.join(SERIALIZED, "Grocery_Store_Status.pkl"))
df_grocery.columns = [col.upper() for col in df_grocery.columns]
```

```python
# filter out irrelevant rows
df_grocery = df_grocery.loc[df_grocery["NEW STATUS"] == "OPEN"]
df_grocery = df_grocery.loc[df_grocery["LOCATION"].notnull()]
```

```python
# format geoloaction
for idx, row in df_grocery.iterrows():
    loc_split = row.LOCATION.split()
    lon = loc_split[1].strip("()")
    lat = loc_split[2].strip("()")
    df_grocery.loc[idx, "LATITUDE"] = lat
    df_grocery.loc[idx, "LONGITUDE"] = lon

# fix data types
df_grocery["LATITUDE"] = df_grocery["LATITUDE"].astype(float)
df_grocery["LONGITUDE"] = df_grocery["LONGITUDE"].astype(float)
```

```python
# drop unnecessary columns
df_grocery.drop(["LAST UPDATED", "LOCATION"], axis=1, inplace=True)
```

```python
# Create a Map instance for Chicago
m = folium.Map(location=[41.8781, -87.6298], zoom_start=10)

for idx, row in df_grocery.iterrows():
    map_functions.add_map_marker(m, row.LATITUDE, row.LONGITUDE, row["STORE NAME"])
#     add_map_circle(m, row.Latitude, row.Longitude)

# Change the map style
# m.add_tile_layer(tiles='Stamen Toner', name='Stamen Toner')

# Display the map
m
```
