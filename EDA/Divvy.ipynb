```python
df = pd.read_pickle(os.path.join(SERIALIZED, "Divvy_Bicycle_Stations_-_In_Service.pkl"))
```

```python
# drop unnecessary columns
df.drop([
    "ID",
    "Total Docks",
    "Docks in Service",
    "Location"
], axis=1, inplace=True)
```

```python
# filter out stations that are south 51st
df = df.loc[df.Latitude > 41.80080416095567]

# filter out stations that are north of Edgewater
df = df.loc[df.Latitude < 42.004912908131764]
```
